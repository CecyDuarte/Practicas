```r
indices<-read.csv(file="indices_bursatiles_retornos _2020.csv", header=TRUE, sep=",", dec=".")
```

```r
plot.ts(indices)
```

```r
indices_tras=t(indices)
indices_tras<-indices_tras[2:7,]
indices_tras
```

```r
library(dtw)
```

**Creación de dendograma**

```r
modelo_dtw<-hclust(dist(indices_tras, method="dtw"))
```

```r
plot(modelo_dtw)
```

```r
library(graphics)

modelo_dtw <- hclust(dist(indices_tras, method="dtw"))
plot(modelo_dtw)
rect.hclust(modelo_dtw, k = 3, border = "red") # Le indicamos la cantidad de clusters que queremos obtener
x <- rect.hclust(modelo_dtw, h = 50, which = c(2,7), border = 3:4)
x
```

```r
Clusters_dtw<-cutree(modelo_dtw, k=3)
Clusters_dtw
```

```r
indices_con_clusters<-cbind(indices_tras, Clusters_dtw)
indices_con_clusters
```
